```python
#now let us try to use different thresholds for binarization

#I am going to define my own binaraization function.

def binarize(image,threshold):
    #convert the image to grey scale first
    result = image.convert('L')
    
    for i in range(result.height):
        for j in range(result.width):
            if result.getpixel((j,i))<threshold:
                result.putpixel((j,i),0)
            else: 
                result.putpixel((j,i),255)
    return result

```

```python
for thresh in range(32,256,32):
    print('threshold = ', thresh)
    print(pytesseract.image_to_string(binarize(img,thresh),'eng+sin'))
    print('------------------------------------------------------------')
    
```
